**CNN**



```python

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

Train CNN by fetching DRIVE **file**

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
train_set = train_datagen.flow_from_directory(
    'dataset/training_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)
test_datagen = ImageDataGenerator(rescale=1./255)

test_set = test_datagen.flow_from_directory(
    'dataset/test_set',  # path to test folder
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary',
    shuffle=False   # important for evaluation
)
```

Building CNN

```python
cnn=tf.keras.models.Sequential()  #Builds a model
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))  # Builds Convolution layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))  # Builds Convolution layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
cnn.add(tf.keras.layers.Flatten())
```

Fully Connected Layer

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))#Just before Output
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))#Outputlayer
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np
```

Now **Compile**

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
cnn.fit(x=train_set,validation_data=test_set,epochs=15)

```

```python
test_image=image.load_img('dataset/single_prediction/cat_or_dog_1.jpg',target_size=(64,64))
test_image=image.img_to_array(test_image)
test_image=np.expand_dims(test_image,axis=0)
result=cnn.predict(test_image)
train_set.class_indices
print(result)
if result[0][0]>0.5:
  prediction = 'dog'
else:
  prediction = 'cat'
```
